```python
df = pd.read_csv('train.csv')
```

```python
# Descrição estatística das variáveis numéricas
df.describe()
```

```python
# Descrição estatística das variáveis categóricas
df.describe(include='object')
```

```python
# Checar valores faltantes
df.isnull().sum()
```

```python
# Visualizar valores faltantes com um heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
# Boxplot para detectar outliers nas variáveis numéricas
num_cols = df.select_dtypes(include=[np.number]).columns

# Calcular o número de linhas e colunas para os subplots
n_cols = 3
n_rows = (len(num_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 5))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(col)

# Remover subplots vazios
for j in range(i + 1, n_rows * n_cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Formatar apenas a tabela de variáveis numéricas
formatted_numeric_desc = df.describe().style.format("{:.2f}")
formatted_numeric_desc
```

```python
# Para variáveis categóricas, não aplicar formatação numérica
categorical_desc = df.describe(include='object')
categorical_desc
```

```python
# Distribuição das variáveis numéricas
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Distribuição das variáveis categóricas
cat_cols = df.select_dtypes(include=['object']).columns
n_cols = 3
n_rows = (len(cat_cols) + n_cols - 1) // n_cols

plt.figure(figsize=(n_cols * 5, n_rows * 5))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Matriz de correlação
corr_matrix = df.corr()
```

```python
# Heatmap da matriz de correlação
plt.figure(figsize=(25, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlação')
plt.show()
```
